```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# 1. ファイルの読み込み（ファイル名は自分のものに変えてください）
file_path = '/content/finalyakisoba.mp3'
y, sr = librosa.load(file_path)

# 2. 特徴量の抽出（音量 RMS）
rms = librosa.feature.rms(y=y)[0]
times = librosa.times_like(rms, sr=sr)

# 3. 盛り上がり（笑い声）の判定
# 全体の平均音量の2倍以上を「盛り上がり」とする例
threshold = np.mean(rms) * 2
peaks = times[rms > threshold]

# 4. 結果の表示
print(f"検知された盛り上がり地点の数: {len(peaks)}")
if len(peaks) > 0:
    print(f"最初の盛り上がり: {peaks[0]:.2f}秒")

# 5. グラフの表示（これを見てしきい値を調整します）
plt.figure(figsize=(10, 4))
plt.plot(times, rms, label='Volume')
plt.axhline(y=threshold, color='r', linestyle='--', label='Threshold')
plt.legend()
plt.show()
```
